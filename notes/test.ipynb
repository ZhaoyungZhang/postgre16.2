```python
import matplotlib.pyplot as plt

# 线程数
threads = [1, 4, 8, 16, 32]

# baseline执行时间范围
baseline_times = [47.31, 196.53, 416.65, 1517.33, 2109.45]

# state recovery执行时间范围
state_recovery_times = [0.11, 0.19, 0.24, 0.33, 0.42]

plt.plot(threads, baseline_times, marker='o', label='Baseline')
plt.plot(threads, state_recovery_times, marker='s', label='State Recovery')

plt.xlabel('Number of Threads')
plt.ylabel('Execution Time (ms)')
plt.title('Baseline vs State Recovery Execution Time')
plt.xticks(threads)
plt.yscale('log')
plt.legend()
plt.grid(True)

for i, txt in enumerate(baseline_times):
    plt.text(threads[i], baseline_times[i], str(txt), ha='right')
for i, txt in enumerate(state_recovery_times):
    plt.text(threads[i], state_recovery_times[i], str(txt), ha='right')

plt.show()
```
